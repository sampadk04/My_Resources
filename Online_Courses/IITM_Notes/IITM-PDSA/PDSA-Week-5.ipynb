```python
# This is an implementation of Kruskal


class XYZ_Courier:
    def __init__(self, AList):
        self.rmap = AList  # Route Map

    def Djikstra(self, s):
        infinity = 1000000001

        vL = self.rmap.keys()
        visited, distance, parent = {}, {}, {}

        for u in vL:
            visited[u], distance[u], parent[u] = False, infinity, -1

        distance[s] = 0
        for _ in vL:
            nextd, nextv = min([(distance[v], v) for v in vL if not (visited[v])])

            visited[nextv] = True
            for (v, d) in self.rmap[nextv]:
                if not (visited[v]):
                    if distance[v] >= distance[nextv] + d:
                        distance[v], parent[v] = distance[nextv] + d, nextv

        return distance, parent

    def cost(self, s, d):
        distance, parent = self.Djikstra(s)
        dist_s_d = distance[d]
        return 5 * dist_s_d

    def route(self, s, d):
        distance, parent = self.Djikstra(s)
        route_s_d = []
        prev_parent = d
        while prev_parent != -1:
            route_s_d.append(prev_parent)
            prev_parent = parent[prev_parent]
        route_s_d.reverse()
        return route_s_d
```

```python
# This is an implementation of Bellman-Ford (Single Source Shortest Paths for -ve Weights) - assuming no -ve cycles are present


def min_cost(r_map, s, d):
    infinity = 1000000001
    distance, parent = {}, {}

    vL = r_map.keys()
    for v in vL:
        distance[v] = infinity
        parent[v] = -1

    distance[s] = 0

    for _ in vL:
        for u in vL:
            for (v, dist) in r_map[u]:
                if distance[v] > distance[u] + dist:
                    distance[v] = distance[u] + dist
                    parent[v] = u

    route_s_d = []
    prev_parent = d
    while prev_parent != -1:
        route_s_d.append(prev_parent)
        prev_parent = parent[prev_parent]
    route_s_d.reverse()

    return distance[d], route_s_d

```

```python
def FiberLink(dmap):
    # Kruskal's Algo Implementation
    edges, comp = [], {}

    l, ans = 0, 0
    for u in dmap.keys():
        edges.extend([(d, u, v) for (v, d) in dmap[u]])
        comp[u] = u
        l += 1
    edges.sort()

    ne = 0
    for (d, u, v) in edges:
        if ne < l:
            if comp[u] != comp[v]:
                ne += 1
                ans += d
                c, d = comp[u], comp[v]
                for w in dmap.keys():
                    if comp[w] == c:
                        comp[w] = d
        else:
            break

    return ans
```

```python
def djikstra(WList, s):
    infinity = 1000000001

    vL = WList.keys()
    visited, distance, parent = {}, {}, {}

    for u in vL:
        visited[u], distance[u], parent[u] = False, infinity, -1

    distance[s] = 0
    for _ in vL:
        nextd, nextv = min([(distance[v], v) for v in vL if not (visited[v])])

        visited[nextv] = True
        for (v, d) in WList[nextv]:
            if not (visited[v]):
                if distance[v] >= distance[nextv] + d:
                    distance[v], parent[v] = distance[nextv] + d, nextv

    return distance, parent


def min_cost_walk(WList, S, D, V):
    distance, parent = djikstra(WList, V)
    mind = distance[S] + distance[D]
    ps, pd = [S], [D]
    u = S
    while parent[u] != -1:
        u = parent[u]
        ps.append(u)
    u = D
    while parent[u] != -1:
        u = parent[u]
        pd.append(u)

    pd.pop()
    pd.reverse()

    p = ps + pd

    return mind, p

```

```python
def IsNegativeWeightCyclePresent(WList):
    # Bellman-Ford Algorithm (to detect -ve weight cycles)
    infinity = 1000000001
    d1, d2 = {}, {}
    vL = WList.keys()
    for v in vL:
        d1[v], d2[v] = infinity, infinity

    s = list(vL)[0]

    for _ in range(len(vL)):
        for u in vL:
            for (v, d) in WList[u]:
                d2[v] = d1[v]
                d1[v] = min(d1[v], d1[u] + d)

    if d1 == d2:
        return False
    else:
        return True
```
