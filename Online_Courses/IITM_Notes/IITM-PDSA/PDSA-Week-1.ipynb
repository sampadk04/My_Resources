```python
def Find_Min_Difference(L,P):
  L.sort()
  n = len(L)
  minimum = 10000
  for i in range(n-P):
    l = L[i:i+P]
    d = l[-1]-l[0]
    if d < minimum:
      minimum = d
  return minimum
```

```python
def isprime(n):
  if n == 1:
    return False
  elif n == 2:
    return True
  else:
    for i in range(2,n//2+1):
      if n%i == 0:
        return False
    return True

def Goldbach(n):
  op = []
  for x in range(2,n//2+1):
    if isprime(x) and isprime(n-x):
      op.append((x,n-x))
  return op
```

```python
def odd_one(L):
  tl = [int, float, str, bool]
  tls = ["int", "float", "str", "bool"]
  d = {}
  for x in L:
    t = type(x)
    it = tl.index(t)
    if it in d:
      d[it] += 1
    else:
      d[it] = 1
  for it in d:
    if d[it] == 1:
      return tls[it]
```

```python
def isprime(n):
  if n == 1:
    return False
  elif n == 2:
    return True
  else:
    for i in range(2,n//2+1):
      if n%i == 0:
        return False
    return True

def primeproduct(m):
    for i in range(2,m//2+1):
        if isprime(i) and (m%i==0) and isprime(m//i):
            return True
    return False

primeproduct(6)
```

```python
def delchar(s,c):
    if len(c) != 1:
        return s
    s = list(s)
    while(s.count(c)>0):
        s.remove(c)
    return(''.join(s))

delchar("banana","a")
```

```python
def shuffle(l1,l2):
    i,j = 0,0
    n,m = len(l1),len(l2)
    opl = []
    while (i<n) and (j<m):
        opl.append(l1[i])
        opl.append(l2[j])
        i += 1
        j += 1
    if n!=m:
        if n>m:
            opl = opl + l1[i:]
        else:
            opl = opl + l2[j:]
    return opl

shuffle([0],[1,3,5])
```

```python
def expanding(L):
    if len(L)<2:
        return True
    mark = abs(L[1]-L[0])
    for i in range(1,len(L)-1):
        if mark >= abs(L[i+1]-L[i]):
            return False
        else:
            mark = abs(L[i+1]-L[i])
    return True

expanding([1,3,7,2,9])
```

```python
def sumsquare(l):
    o = [x**2 for x in l if (x%2 == 1)]
    e = [x**2 for x in l if (x%2 == 0)]
    return[sum(o),sum(e)]

sumsquare([1,3,5])
```

```python
class Open_Book:
	def __init__(self, d):
		self.d = d
		self.words = list(d.values())
	
	def bs_word(self, w, s, e):
		l = self.words
		if s==e :
			pg = s//5
			ps = s%5
			if l[pg][ps] == w:    
				return (pg+1,ps+1)
			else:
				return (-1,-1)
		if s>e:
			return (-1,-1)

		m = (s+e)//2
		pg = m//5
		ps = m%5
		if w == l[pg][ps]:
			return (pg+1,ps+1)		
		if w < l[pg][ps]:
			return self.bs_word(w,s,m-1)
		else:
			return self.bs_word(w,m+1,e)
	
	def search_word(self, w):
		l = self.words
		n = len(l)*5
		(x,y) = self.bs_word(w,0,n-1)
		if x == -1:
			return "Not found"
		else:
			return (x,y)
```

```python
def merge(A,B):
    m,n = len(A),len(B)
    C,i,j,k = [],0,0,0
    while k < m+n:
        if i == m:
            C.extend(B[j:])
            k = m+n
        elif j == n:
            C.extend(A[i:])
            k = m+n
        elif A[i][1] == B[j][1]:
            C.extend([A[i],B[j]])
            i,j,k = i+1,j+1,k+2
        elif A[i][1] < B[j][1]:
            C.append(A[i])
            i,k = i+1,k+1
        else:
            C.append(B[j])
            j,k = j+1,k+1
    return C
```

```python
def histogram(l):
    s = list(set(l))
    ol1 = []
    for x in s:
        ol1.append((l.count(x),x))
    ol1.sort()
    ol2 = [(b,a) for (a,b) in ol1]
    return ol2
```
