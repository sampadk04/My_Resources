```python
def DishPrepareOrder(l):
    id = set()
    idc = []
    for x in l:
        if not(x in id):
            idc.append((1,(-1)*x))
            id.add(x)
        else:
            for i in range(len(idc)):
                if idc[i][1] == (-1)*x:
                    (a,b) = idc[i]
                    idc[i] = (a+1,b)
                    break

    idc.sort(reverse=True)

    op = [(-1)*y for (x,y) in idc]

    return op

l = [1006,1008,1009,1008,1007,1005,1008,1001,1003,1009,1006,1003,1004,1002,1008,1005,1004,1007,1006,1002,1002,1001,1004,1001,1003,1007,1007,1005,1004,1002]

s = DishPrepareOrder(l)

print(s)
```

```python
def eval(a,b,x):
    op = ['+','-','*','/','**']
    i = op.index(x)
    if i == 0:
        return a+b
    elif i == 1:
        return a-b
    elif i == 2:
        return a*b
    elif i == 3:
        return a/b
    else:
        return a**b

def EvaluateExpression(e):
    e = e.split()
    op = ['+','-','*','/','**']
    ans = []
    for x in e:
        if x in op:
            b = ans.pop()
            a = ans.pop()
            ans.append(eval(a,b,x))
        else:
            ans.append(float(x))
    
    fans = ans.pop()
    return fans

s = "2 3 1 * + 9 -"
EvaluateExpression(s)
```

```python
def recRev(prev,curr,index=0):
    if curr.next==None:
        curr.next=prev
        return curr
    else:
        a=recRev(curr,curr.next,index+1)
        if(index==0):
            curr.next=prev
            prev.next=None
        else:
            curr.next=prev
        return a
        
    

def reverse(root):
    if root.next==None:
        return root
    return recRev(root,root.next)
```
