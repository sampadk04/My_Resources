# Task 2

1. Generate a classification dataset using `make_classification` using the following details:
    - 500 samples
    - 2 features
    - 2 classes
    - one cluster per class

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0, n_classes=2, n_clusters_per_class=1)

# NOTE: Number of informative, redundant and repeated features must sum to less than the number of total features
# This means n_features >= n_informative + n_redundant + n_repeated
# So, n_features = 2, we can't have the default value of n_redundant = 2, we have to change to something lower, i.e. 0
```

```python
print(X.shape, y.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
plt.scatter(X[:,0],X[:,1], c=y, marker = '+')
# Here c = y (output vector) is showing us the class in which each point in the scatterplot belongs to (each point here corresponds to an input point with a label). Here, yellow represents the points in class 0 and purple represents points in class 1.
# c = y, is different from something like a c = r, which basically fixes the colour of the point to red.
```

# Task 7

1. Generate and visualize the binary classification data using `numpy`.
2. Generate 25 samples of `class_0` with 2 features.
3. Generate 25 samples of `class_1` with 2 features.

```python
# This gives 50 normally distributed data in a numoy array.
np.random.normal(size = 50)
```

```python
# This basically reshapes 50 normally generated numbers into 25*2 numpy matrix
class_0_data = np.random.normal(size = 50).reshape(25,2) - 10
class_0_data
```

```python
class_1_data = np.random.normal(size = 50).reshape(25,2) + 10
class_1_data
```

```python
X = np.concatenate([class_0_data,class_1_data])
X
```

```python
class_0_labels = np.zeros(25)
class_1_labels = np.ones(25)
y = np.concatenate([class_0_labels,class_1_labels])
y
```

```python
plt.scatter(X[:,0],X[:,1], c=y, marker='+')
```
