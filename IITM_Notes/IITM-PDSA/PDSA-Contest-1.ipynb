```python
T = int(input())

for _ in range(T):
    N,K = map(int,input().split())
    
    if K>N:
        print(N)
    elif (K >= N//2 + 1):
        if (N%2 == 0):
            print(N-N//2-1)
        else:
            print(N - (N+1)//2)
    else:
        maxr = -1
        for i in range(1,K+1):
            r = N%i
            if r>maxr:
                maxr = r
        print(maxr)
```

```python
T = int(input())

for _ in range(T):

    N = int(input())
    S = list(map(int,input().split()))

    S.sort()
    mins = 10000000000

    for i in range(len(S)-1):
        d = abs(S[i]-S[i+1])
        if d < mins:
            mins = d
    
    print(mins)
```

```python
def isRainbow(N,L):
    if max(L) > 7:
        return False
    else:
        lc = [0 for i in range(8)]
        for i in range(N):
            lc[L[i]] += 1
        for i in range(1,7):
            if lc[i] == 0 or lc[i] % 2 == 1:
                return False
        if lc[7] == 0:
            return False
    
    m = N//2
    l,r = L[:m],L[m:]
    r.reverse()
    if N%2 == 1:
        l.append(7)
    
    curr = 1
    
    for i in range(len(l)):
        if l[i] != r[i]:
            return False
        
        if l[i] != curr:
            if l[i] != curr + 1:
                return False
            else:
                curr = L[i]
    
    return True


T = int(input())

for _ in range(T):

    N = int(input())
    L = list(map(int,input().split()))

    if isRainbow(N,L):
        print("yes")
    else:
        print("no")
```

```python
T = int(input())

for _ in range(T):

    N,K = map(int,input().split())
    L = list(map(int,input().split()))
    L.sort()
    
    K = min(K,N-K)
    s1 = sum(L[:K])
    s2 = sum(L[K:])

    print(s2-s1)
```

```python
T = int(input())

for _ in range(T):

    n,k = map(int,input().split())
    L = list(map(int,input().split()))
    L.sort()

    m = L[-1]
    r = m+1-n

    lc = [0 for _ in range(m+1)]
    for x in L:
        lc[x] += 1

    if k>=r:
        print(m+1+k-r)
    else:
        if k == 0:
            i = 0
            while lc[i] == 1:
                i += 1
            print(i)
        else:
            i,k = 0,k+1
            while k>0:
                if lc[i] == 0:
                    k -= 1
                    i += 1
                else:
                    i += 1
            print(i-1)
```

```python
N,W,L = map(int,input().split())

H,R = [],[]

for _ in range(N):
    h,r = map(int,input().split())
    H.append(h)
    R.append(r)

m = 0
Ha = [x for x in H if (x>=L)]
s = sum(Ha)

while s<W:
    m += 1
    Ha = []
    for i in range(N):
        H[i] += R[i]
        if H[i] >= L:
            Ha.append(H[i])
    s = sum(Ha)

print(m)
```
