```python
def isort(l):
  n = len(l)
  if n<1:
    return l
  for i in range(n):
    j = i
    while(j>0 and l[j][0] < l[j-1][0]):
      l[j],l[j-1] = l[j-1],l[j]
      j = j-1
  return l

def disort(l):
  n = len(l)
  if n<1:
    return l
  for i in range(n):
    j = i
    while(j>0 and l[j][0] == l[j-1][0] and int(l[j][1:])>int(l[j-1][1:])):
      l[j],l[j-1] = l[j-1],l[j]
      j = j-1
  return l

def combinationSort(l):
  l1 = isort(l)
  l2 = l1[:]
  l2 = disort(l2)
  return (l1,l2)
```

```python
def findLargest(l):
    n = len(l)
    if n==1:
        return l[0]
    elif l[0] < l[-1]:
        return l[-1]
    else:
        mid = n//2
        if l[mid] <= l[-1]:
            return findLargest(l[:mid])
        else:
            return findLargest(l[mid:])

print(findLargest([1,2,3,4]))
print(findLargest([4,1,2,3]))
print(findLargest([3,4,1,2]))
print(findLargest([2,3,4,1]))
print(findLargest([0]))
```

```python
def mergeInPlace(A,B):
    i = 0
    while(i<len(A)):
        if A[i] <= B[0]:
            i += 1
        else:
            A.swap(i,B,0)
            i += 1
            resort(B)

def resort(L):
    n = len(L)
    if n>1:
        for i in range(n):
            j = i
            while(j>0 and L[j]<L[j-1]):
                B.swap(j,B,j-1)
                j -= 1
```

```python
def bs(l,x):
    if l == []:
        return False
    
    m = len(l)//2

    if x == l[m]:
        return True
    elif x < l[m]:
        return bs(l[:m],x)
    else:
        return bs(l[m+1:],x)

def findIntersection(l1,l2):
    m,n = len(l1),len(l2)
    if m<n:
        m,n,l1,l2 = n,m,l2,l1
    intl = []
    l2.sort()
    for x in l1:
        if bs(l2,x):
            intl.append(x)
    return intl

l1 = [5,8,2]
l2 = [6,8]

findIntersection(l1,l2)
```
